```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('daTA.csv')
df.head()
```

```python
df.info()
```



*  **Define a CarSale Class (OOP Concept)**




```python
class CarResale:
  def __init__(self, brand, model, year, price):
    self.brand = brand
    self.model = model
    self.year = year
    self.price = price

  def car_age(self, current_year = 2025):
    return current_year - self.year

  def is_luxuary(self):
    return self.price > 2000000
```

```python
example  = CarResale("Toyota","Camry",2018,160000)
print(f"Example Car is {example.car_age()} years old. Luxuary:{example.is_luxuary()}")
```

Basic Statistics

```python
def basic_analysis(self):
  print("\n First 5 Records")
  print(self.df.head())

  print("\n Column Info:")
  print(self.df.info())

  print("\n Summary Stastitics:")
  print(self.df.describe())

  print("\n Fuel Type Distribution:")
  print(self.df['Fuel_type'].value_counts())
```

NumPy Use – Price Range & Outliers

```python
def price_range_insight(self):
  prices = self.df['Selling_Price'].values
  print("\n Price Range:{np.min(prices)} to {np.max(prices)}")
  print(self.df[self.df['Selling_Price'] > np.percentile(prices, 95)])
```

```python
df = df.dropna()
```

Feature Engineering With Numpy

```python
#New Column = Age Of Car
current_year = 2025
df['Car_Age'] = current_year - df['Year']
```

```python
# Classify price segments using NumPy
price_array = df['Selling_Price'].values # Corrected column name to 'Selling_Price'
price_segment = np.where(price_array > 2000000, 'Luxury', 'Standard')
df['Segment'] = price_segment
```

```python
df
```

Descriptive Statistics Using Pandas

```python
print("\n Descriptive Statistics:")
print(df.describe())
```

```python
df.rename(columns= {
    'Car_Name' : 'Brand',
    'Present_Price' : 'Price'
    }, inplace = True)
```

```python
#Avg Price Brand Wise

print("\nAverage Price per Brand:")
print(df.groupby('Brand')['Price'].mean())
```

Data Visulization

```python
sns.set(style="whitegrid")
```

```python
# Histogram of Car Prices
plt.figure(figsize=(8,5))
plt.hist(df['Price'], bins = 20, color='Skyblue', edgecolor = 'black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
# Bar Plot

plt.figure(figsize=(10,6))
sns.barplot(data = df, x = 'Brand', y = 'Price', estimator=np.mean)
plt.title('Average Price per Brand')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Box Plot: Price Vs Segment

plt.figure(figsize=(8,5))
sns.boxplot(data = df, x = 'Segment', y = 'Price')
plt.title('Price Distribution by Segment')
plt.show()
```

```python
# Scatter Plot : Price Vs Age Car

plt.figure(figsize=(8,5))
sns.scatterplot(data = df, x = 'Car_Age', y = 'Price', hue = 'Segment')
plt.title('Price vs Age of Car')
plt.show()
```
